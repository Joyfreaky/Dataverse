```python
# Encode the object data types to numeric data types
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Encode the train data expect the target variable and the ID variable
for col in train.columns:
    if train[col].dtype == 'object' and col not in ['policy_id', 'is_claim']:
        train[col] = le.fit_transform(train[col])


# Encode the test data
for col in test.columns:
    if test[col].dtype == 'object' and col not in ['policy_id']:
        test[col] = le.fit_transform(test[col])
        

```

```python
# Check the data types of the train data
train.dtypes
```

```python
# if the column has only two unique values, then it is a binary variable
binary_cols = [col for col in train.columns if train[col].nunique() == 2]

# if the column has more than two unique values, then it is a categorical variable
categorical_cols = [col for col in train.columns if train[col].nunique() > 2]

# Check the binary columns
print(binary_cols)

# Check the categorical columns
print(categorical_cols)
```

```python
# Convert the binary columns to binary data types
for col in binary_cols:
    train[col] = train[col].astype('bool')
    # test except the target variable
    if col not in ['is_claim']:
        test[col] = test[col].astype('bool')



# Check the data types of the train data
train.dtypes
```

```python
# Check the data types of the test data
test.dtypes
```

```python
# Convert the categorical columns to categorical data types
for col in categorical_cols:
    # expect the policy_id
    if col not in ['policy_id']:
        train[col] = train[col].astype('category')
        test[col] = test[col].astype('category')
    

```

```python
# Change the data type of the is_claim column to category
train['is_claim'] = train['is_claim'].astype('category')
```

```python
# Change False to 0 and True to 1
train['is_claim'] = train['is_claim'].cat.codes

```

```python
# Change the is_claim column to category
train['is_claim'] = train['is_claim'].astype('category')
```

```python
# Build decision tree model
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Import library to check F1 score
from sklearn.metrics import f1_score


# Split the train data into train and validation set
# Use is_claim as the target variable
# Use all the other variables as the features
# Drop the policy_id variable
# Set the random state to 42
X_train, X_val, y_train, y_val = train_test_split(train.drop(['policy_id', 'is_claim'], axis=1), train['is_claim'], test_size=0.2, random_state=42)

# Build the model
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the validation set
preds = model.predict(X_val)

# Calculate the F1 score with the validation set
print(f1_score(y_val, preds))

# Check the feature importance
model.feature_importances_




```

```python
# Make predictions on the test set
preds_test = model.predict(test.drop(['policy_id'], axis=1))

# Import the sample submission file
sample_submission = pd.read_csv('/home/joydipb/Documents/Dataverse/data/sample_submission.csv')

# Create a dataframe with the policy_id and the predictions
submission = pd.DataFrame({'policy_id': test['policy_id'], 'is_claim': preds_test})

# Check the format of the sample submission file
sample_submission.head()
```

```python
# Create a Light GBM model
from lightgbm import LGBMClassifier

# Build the model
model = LGBMClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the validation set
preds = model.predict(X_val)

# Calculate the F1 score with the validation set
print(f1_score(y_val, preds))
```

```python
# AutoML with H2O
import h2o
from h2o.automl import H2OAutoML

# Start the H2O server
h2o.init()

# Convert the train data to H2OFrame
train_h2o = h2o.H2OFrame(train)

# Convert the test data to H2OFrame
test_h2o = h2o.H2OFrame(test)

# Set the target variable
y = 'is_claim'

# Set the features epcept the target variable and the policy_id
x = train_h2o.columns
x.remove(y)
x.remove('policy_id')

# Build the model
aml = H2OAutoML(max_models=20, seed=42)
aml.train(x=x, y=y, training_frame=train_h2o)

# Get the leaderboard
lb = aml.leaderboard

# Check the leaderboard
lb.head()
```

```python
# Choose the best model
model = aml.leader

# Make predictions on the test set and use threshold of 0.5
preds = model.predict(test_h2o).as_data_frame()['predict'].apply(lambda x: 1 if x > 0.05 else 0)

# Create a dataframe with the policy_id and the predictions
submission = pd.DataFrame({'policy_id': test['policy_id'], 'is_claim': preds})

# Check the first five rows of the submission file
submission.head()




```
